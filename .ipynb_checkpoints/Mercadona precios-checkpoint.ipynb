```python
from selenium import webdriver
from bs4 import BeautifulSoup
import csv

# Configure Selenium
driver = webdriver.Chrome()
driver.get('https://tienda.mercadona.es/categories/77')

# Wait for the page to load completely
import time
time.sleep(5)  # Adjust as necessary

# Get HTML from the loaded page
html = driver.page_source

# Use BeautifulSoup to analyze HTML
soup = BeautifulSoup(html, 'html.parser')
# Update classes here
egg_names = soup.find_all('h4', class_='subhead1-r product-cell__description-name')
egg_sizes = soup.find_all('div', class_='product-format product-format__size--cell')
egg_prices = soup.find_all('p', class_='product-price__unit-price subhead1-b')

# Check if data were fetched
print(f"Eggs: {len(egg_names)}, Prices: {len(egg_prices)}, Sizes: {len(egg_sizes)}")

# Close browser
driver.quit()

# Save data in CSV file
if egg_names and egg_sizes and egg_prices:
    with open('eggs.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Egg', 'Size', 'Price'])
        for egg_name, egg_size, egg_price in zip(egg_names, egg_sizes, egg_prices):
            writer.writerow([egg_name.text.strip(), egg_size.text.strip(), egg_price.text.strip()])
           
    print("Data saved in eggs.csv")
else:
    print("No product or price found.")
```

# tentando escrever o codigo inteiro das frutas

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from bs4 import BeautifulSoup
import csv
import time

# Configure Selenium
driver = webdriver.Chrome()
driver.get('https://tienda.mercadona.es/categories/27')

# Wait for the page to load completely
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '.product-price__unit-price.subhead1-b')))

# Get HTML from the loaded page
html = driver.page_source

# Use BeautifulSoup to analyze HTML
soup = BeautifulSoup(html, 'html.parser')


# Lists to store data
fruit_names = []
fruit_sizes = []
fruit_prices = []

products = soup.find_all('div', class_="product-cell product-cell--actionable")

for product in products:
    # Get the name and size
    fruit_name = product.find('h4').text.strip()
    fruit_size = product.find('div', class_='product-format product-format__size--cell').text.strip()

    # Try to find the price
    price_tag = product.find('p', class_='product-price__unit-price subhead1-b')
    fruit_price = price_tag.text.strip() if price_tag else 'N/A'  # 'N/A' if no price found

    
    fruit_names.append(fruit_name)
    fruit_sizes.append(fruit_size)
    fruit_prices.append(fruit_price)

# Print the results

print(f"Fruits: {len(fruit_names)}, Prices: {len(fruit_prices)}, Sizes: {len(fruit_sizes)}")

for fruit_name, fruit_size, fruit_price in zip(fruit_names, fruit_sizes, fruit_prices):
    print(f'Fruit: {fruit_name}, Size: {fruit_size}, Price: {fruit_price}')
    
    
# Save data in a CSV file if matches found
if fruit_names and fruit_sizes and fruit_prices:
    with open('fruits.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Fruit', 'Price', 'Size'])
        for fruit_name, fruit_size, fruit_price in zip(fruit_names, fruit_sizes, fruit_prices):
            writer.writerow([fruit_name, fruit_size, fruit_price])
    print("Data saved in fruits.csv")
else:
    print("No fruit, price, or size found.")
```

```python
import pandas as pd

# Abre o CSV e transforma em DataFrame
df = pd.read_csv('fruits.csv')

# Exibe as primeiras linhas do DataFrame
print(df.head())
```

```python
import pandas as pd

# Extrair o número de unidades da coluna 'Size'
df_eggs['Units'] = df_eggs['Size'].str.extract(r'(\d+)\s*ud')[0].astype(float)

# Limpar a coluna 'Price' removendo o símbolo do euro e converter para float
df_eggs['Price'] = df_eggs['Price'].str.replace('€', '').str.replace(',', '.').astype(float)

# Calcular o preço por unidade
df_eggs['PricePerUnit'] = df_eggs['Price'] / df_eggs['Units']

df_eggs
```

```python
import pandas as pd

df_fruits = pd.read_csv('fruits.csv')

# Limpar a coluna 'Price' removendo o símbolo do euro e convertendo para float
df_fruits['Price'] = df_fruits['Price'].str.replace('€', '').str.replace(',', '.').astype(float)

# Extrair o peso da coluna 'Size' e converter para quilogramas
def extract_weight(size):
    size = size.lower()
    if 'kg' in size:
        return float(size.split()[0].replace(',', '.'))
    elif 'g' in size:
        return float(size.split()[0]) / 1000
    else:
        return float('nan')

df_fruits['Weight'] = df_fruits['Size'].apply(extract_weight)

# Calcular o preço por quilo
df_fruits['PricePerKg'] = df_fruits['Price'] / df_fruits['Weight']

# Criar DataFrame para bananas/plátanos
platanos_df = df_fruits[df_fruits['Fruit'].str.contains('banana|plátano', case=False, na=False)]

# Criar DataFrame para maçãs (manzanas)
manzanas_df = df_fruits[df_fruits['Fruit'].str.contains('manzana', case=False, na=False)]

# Exibir as primeiras linhas dos DataFrames
print("Bananas/Plátanos:")
print(platanoss_df.head())

print("\nManzanas:")
print(manzanas_df.head())
```
