```python
import pandas as pd
df_limonc = pd.read_csv("limoncello.csv")

def limoncello(value): 
    if value.endswith("cl"):
        cant = int(value.replace("cl", "").strip())
        return str(cant * 10) + " ml"
    else:
        return value

df_limonc['Cantidad'] = df_limonc["Cantidad"].map(limoncello)
df_limonc = df_limonc.head(10)
df_limonc
```

```python
import pandas as pd
df_cost = pd.read_csv("costilla.csv")
df_cost
```

```python
df_cost.drop(df_cost[df_cost['Name'].str.contains("Frijoles")].index)
df_cost[~df_cost['Name'].str.contains("Frijoles")]
df_cost = df_cost[df_cost['Name'].str.contains("Costilla")].reset_index(drop=True)
df_cost
```

```python
def costilla(value): 
    if value.endswith("kg"):
        cant = float(value.replace("kg", "").strip())
        return str(int(cant * 1000)) + " g"
    else:
        return value

df_cost['Cantidad'] = df_cost["Cantidad"].map(costilla)
df_cost
```

```python
import pandas as pd
df_agua = pd.read_csv("agua mineral natural.csv")
df_agua
```

```python
df_agua = df_agua[df_agua['Name'].str.contains("Agua mineral")].reset_index(drop=True)
df_agua
```

```python
def agua(value): 
    if value.endswith("litros"):
        cant = float(value.replace("litros", "").replace(",", ".").strip())
        return str(int(cant * 1000)) + " mL"
    elif value.endswith("cl"):
        cant = float(value.replace("cl", "").strip())
        return str(int(cant * 10)) + " mL"
    else:
        return value

df_agua['Cantidad'] = df_agua["Cantidad"].map(agua)
df_agua
```

```python
df_agua = df_agua.head(10)
df_agua
```

```python
dfs = [df_agua, df_pollo, df_cost, df_huevos, df_aceite, df_leche, df_cafe, df_bolsas, df_arroz, df_pan]
```

```python
productos_mas_baratos = []

for df_produto in dfs:

    menor_precio = df_produto.iloc[0]
    
    productos_mas_baratos.append({
        "Producto": menor_precio["Name"],
        "Supermercado": menor_precio["Super"],
        "Precio (€)": menor_precio["Price (€)"],
        "Precio Uniforme (€)": menor_precio["Uniform Price (€)"]
    })

df_mas_baratos = pd.DataFrame(productos_mas_baratos)

print(df_mas_baratos)
```

```python
df_mas_baratos
```

```python
df_mas_baratos = df_mas_baratos.groupby('Producto', as_index=False).apply(lambda x: x.loc[x['Precio Uniforme (€)'].idxmin()])

frecuencia_supermercados = df_mas_baratos['Supermercado'].value_counts()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.barplot(x=frecuencia_supermercados.index, y=frecuencia_supermercados.values, palette="viridis")

# Adiciona títulos e rótulos
plt.title("Frecuencia de cada Supermercado como el Más Barato")
plt.xlabel("Supermercado")
plt.ylabel("Frecuencia (Cantidad de Productos con el Precio Más Bajo)")

# Exibe o gráfico
plt.show()
```

```python
avg_dfs = []

for df in dfs:
    avg_df = df.groupby("Super", as_index=False)["Uniform Price (€)"].mean()
    avg_df.rename(columns={"Uniform Price (€)": "Average Uniform Price (€)"}, inplace=True)
    avg_dfs.append(avg_df)

avg_dfs
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que avg_dfs es tu lista de DataFrames con la media de precios
df_combined = pd.concat(avg_dfs, ignore_index=True)

# Asegúrate de que tus DataFrames tienen las columnas correctas
# Puedes imprimir df_combined para verificar
print(df_combined.head())

# Organizar para tener un producto por columna y un supermercado por fila
heatmap_df = df_combined.pivot(index="Super", columns="Producto", values="Average Uniform Price (€)")

# Gerar o heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_df, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'Precio Uniforme Medio (€)'})
plt.title("Mapa de Calor de Media de Precios por Producto y Supermercado")
plt.xlabel("Producto")
plt.ylabel("Supermercado")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
