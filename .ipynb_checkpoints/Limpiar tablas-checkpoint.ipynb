```python
import pandas as pd
df = pd.read_csv("limoncello.csv")

def limoncello(value): 
    if value.endswith("cl"):
        cant = int(value.replace("cl", "").strip())
        return str(cant * 10) + " ml"
    else:
        return value

df['Cantidad'] = df["Cantidad"].map(limoncello)
df
```

```python
import pandas as pd
df1 = pd.read_csv("costilla.csv")
df1
```

```python
#df1.drop(df1[df1['Name'].str.contains("Frijoles")].index)
#df1[~df1['Name'].str.contains("Frijoles")]
df1 = df1[df1['Name'].str.contains("Costilla")].reset_index(drop=True)
df1
```

```python
def costilla(value): 
    if value.endswith("kg"):
        cant = float(value.replace("kg", "").strip())
        return str(int(cant * 1000)) + " g"
    else:
        return value

df1['Cantidad'] = df1["Cantidad"].map(costilla)
df1
```

```python
import pandas as pd
df2 = pd.read_csv("agua mineral natural.csv")
df2
```

```python
df2 = df2[df2['Name'].str.contains("Agua mineral")].reset_index(drop=True)
df2
```

```python
def agua(value): 
    if value.endswith("litros"):
        cant = float(value.replace("litros", "").replace(",", ".").strip())
        return str(int(cant * 1000)) + " mL"
    elif value.endswith("cl"):
        cant = float(value.replace("cl", "").strip())
        return str(int(cant * 10)) + " mL"
    else:
        return value

df2['Cantidad'] = df2["Cantidad"].map(agua)
df2
```
